### Case Study: Helios Renewables (Η) - Navigating Growth and a Greener Future

**Introduction & Current Landscape:**

Helios Renewables (Η), an established firm in the renewable energy sector, has built a solid reputation on its portfolio of operating solar farms. These existing operations are projected by analysts to generate a steady $50 million in free cash flow (FCF) each year indefinitely. The company operates in a jurisdiction with a corporate tax rate of 30%. Current market conditions indicate a risk-free rate of 3%, and investors typically demand a market risk premium of 7% for investments of this nature.

Helios' CFO, Ben Carter, recently reviewed the firm's financial standing. The equity markets value Helios' shares at a total of $300 million. Alongside this, the company carries $200 million in debt on its books, which currently costs them 6.0% annually before taxes. Based on their current financial structure and market perception, Helios' equity has a beta of 1.20. The debt, while not explicitly traded with a public beta, is estimated by analysts to have a beta of 0.20.

CEO Anya Sharma is now looking to the future and has called a strategy meeting to discuss the company's financial health and a significant new investment opportunity.

Once operational, it's expected to add a perpetual $12 million in annual FCF, starting next year. Ben Carter is evaluating financing options. "If we were to go the traditional route and fund this like our other assets," he muses, "what would that look like?" He's also been reading about Modigliani-Miller and the potential benefits of debt due to taxes. "Alternatively," he adds, "what if we took this opportunity to issue $100 million in new, permanent debt specifically for Project Aurora? How would that change our cost of capital picture, assuming our cost of debt and debt beta remain stable even at that higher leverage?"
